1. Write a Python Program to Find LCM.

```python
def find_lcm(num1,num2):
    '''This function takes two parameters num1 and num2 to find the lcm of the two''' 
    if num1>num2:
        high_value=num1
    else:
        high_value=num2
        
    while(True):
        if (high_value%num1==0) and (high_value%num2==0):
            LCM = high_value
            
            break
            
        high_value+=1
        
    return print(f"So the lcm of {num1} and {num2} is {LCM}")
```

```python
num1,num2=int(input("Enter first number: ")),int(input("Enter second number: "))

find_lcm(num1,num2)
```

2. Write a Python Program to Find HCF.

```python
def find_hcf(num1,num2):
    '''This function takes two parameters num1 and num2 to find the HCF of the two''' 
    HCF=1
    if num1>num2:
        low_value=num2
    else:
        low_value=num1
        
    for i in range(1,low_value+1):
        if (num1%i==0 and num2%i==0):
            HCF=i     
    return print(f"HCF of {num1} and {num2} is {HCF}.")
```

```python
find_hcf(2,4)
```

3. Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal.

```python
def find_binary_octal_hexadecimal(num):
    '''This function takes decimal number as a parameter and returns its binary octal and hexdecimal conversion.'''
    print(f"The binary conversion of {num} is {bin(num)}.")
    print(f"The octal conversion of {num} is {oct(num)}.")
    print(f"The hexadecimal conversion of {num} is {hex(num)}.")
```

```python
num = int(input("Enter a decimal number: "))

find_binary_octal_hexadecimal(num)
```

5. Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations.

```python
def calc(num1, num2):
    '''This function will take two parameters as input and evaluate the addition, subtraction, multiplication or division '''
    operator=int(input("Enter 1 for addition, 2 for subtraction, 3 for multiplication and 4 for division: "))
    
    if operator==1:
        return sum(num1,num2)
    elif operator==2:
        return (num1-num2)
    elif operator==3:
        return (num1*num2)
    elif operator==4:
        return (num1/num2)
    else:
        return "Invalid input"
```

```python
num1,num2=float(input("Enter first number: ")),float(input("Enter second number: "))

calc(num1, num2)
```
