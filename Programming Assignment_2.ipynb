1. Write a Python program to convert kilometers to miles.

```python
def km_mile_conv(length):
    return len/1.609
```

```python
length=int(input("Give length in km to be converted:"))

km_mile_conv(length)
```

2. Write a Python program to convert Celsius to Fahrenheit.

```python
def celsius_fahrenheit_conv(temp_celsius):
    temp_fahrenheit=(temp_celsius*1.8)+32
    print(f"Temperature in Fahrenheit is {temp_fahrenheit}.")
```

```python
temp_celsius = int(input("Enter temperature in Celsius: "))

celsius_fahrenheit_conv(temp_celsius)
```

4. Write a Python program to solve quadratic equation.

```python
def sol_quad(a,b,c):
    
    d=((b**2)-(4*a*c))**0.5
    
    sol1 = (-b-d)/(2*a)
    sol2 = (-b+d)/(2*a)
    
    print(f"The solution of the give quadratic equation is {sol1} and {sol2}.")
```

```python
print("If ax^2 + bx + c is in the form of quadratic equation.")
a = int(input("Value of coefficient a:"))
b = int(input("Value of coefficient b:"))
c = int(input("Value of constantt c:"))

sol_quad(a,b,c)
```

5. Write a Python program to swap two variables without temp variable.

```python
def swap(a,b):
    
    print(f"Initially a is {a} and b is {b}.")
    
    a,b=b,a
    
    print(f"After swapping a is {a} and b is {b}.")
```

```python
a = int(input("Enter value of a:"))

b = int(input("Enter value of b:"))

swap(a,b)
```
